```python
import numpy as np
import matplotlib.pyplot as plt
from keras.models import load_model
from PIL import Image
from functions import predict
```

```python
# Cargar el historial de entrenamiento
history = np.load('training_history.npy', allow_pickle=True).item()
```

```python
# Cargar el conjunto de datos de prueba
data = np.load('data/data.npz')

test_images = data['testX']
test_labels = data['testY']
```

```python
# Cargar el modelo
model = load_model("trained_models/mobilenet_best.h5")
```

- Curva de pérdida durante el entrenamiento:

```python
plt.plot(history['loss'], color='b', label="Trainning loss")
plt.plot(history['val_loss'], color='purple', label="Validation loss")
plt.xlabel('epochs')
plt.ylabel('loss')
plt.title('Loss function')
plt.legend()
```

- Curva de precisión durante el entrenamiento:

```python
plt.plot(history['accuracy'], color='b', label="Trainning accuracy")
plt.plot(history['val_accuracy'], color='purple', label="Validation accuracy")
plt.xlabel('epochs')
plt.ylabel('accuracy')
plt.title('Loss function')
plt.legend()
```

- Evaluación del modelo en el conjunto de prueba:

```python
_, acc = model.evaluate(test_images, test_labels)
print("Testing Accuracy: {:.2f}".format(acc*100))
```
